```python

data = data.drop('Booking_ID',axis=1)
```

```python
data.head()
```

```python
data = data.sort_values(by=['arrival_year', 'arrival_month', 'arrival_date']).reset_index(drop=True)
split_index = round(len(data) * 0.8)
train = data.iloc[:split_index]
test = data.iloc[split_index:]
```

```python
type(train)
train.head()
```

```python
x_train = train.drop('booking_status', axis=1)
y_train = train['booking_status']

print(x_train.head())
print(y_train.head())
```

```python
x_test = test.drop('booking_status', axis=1)
y_test = test['booking_status']

print(x_test.head())
print()
y_test.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
le.classes_
y_test = le.fit_transform(y_test)
le.classes_
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

num_cols = ['lead_time', 'avg_price_per_room']
cat_cols = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first'), cat_cols)
    ],
    remainder='passthrough'
)

# Fit only on training set
x_train = preprocessor.fit_transform(x_train)

# Apply the same fitted transformation to test set
x_test = preprocessor.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
#intialize the model
model = LogisticRegression(solver='liblinear',random_state=42)
```

```python
#Train the model
model.fit(x_train,y_train)
```

```python
#Make Predictions
y_pred_test = model.predict(x_test)
y_pred_train = model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

#Check accuracy 
print("Test Accuracy:", accuracy_score(y_test,y_pred_test))

#Detailed report
print("Test Classification Report:\n", classification_report(y_test, y_pred_test))

#Confution Matrix
print("Test Confution Matrix:\n", confusion_matrix(y_test,y_pred_test))
print()
#Check accuracy 
print("Train Accuracy:", accuracy_score(y_train,y_pred_train))

#Detailed report
print("Train Classification Report:\n", classification_report(y_train, y_pred_train))

#Confution Matrix
print("Train Confution Matrix:\n", confusion_matrix(y_train,y_pred_train))
```

```python
from sklearn.model_selection import GridSearchCV

#Define the grid of parameters
param_grid = { 
    'C': [0.01,0.1,1,10],
    'solver': ['liblinear', 'lbfgs']
}

#setup GridSerachCV
grid_search = GridSearchCV(LogisticRegression(random_state=4), param_grid, cv =5)
grid_search.fit(x_train,y_train)

#Best parameters
print('Best Parameters:', grid_search.best_params_)
```

```python
best_model= grid_search.best_estimator_

y_pr_test = best_model.predict(x_test)
y_pr_train = best_model.predict(x_train)
```

```python
#Check accuracy 
print("Test Accuracy:", accuracy_score(y_test,y_pr_test))

#Detailed report
print("Test Classification Report:\n", classification_report(y_test, y_pr_test))

#Confution Matrix
print("Test Confution Matrix:\n", confusion_matrix(y_test,y_pr_test))
print()
#Check accuracy 
print("Train Accuracy:", accuracy_score(y_train,y_pr_train))

#Detailed report
print("Train Classification Report:\n", classification_report(y_train, y_pr_train))

#Confution Matrix
print("Train Confution Matrix:\n", confusion_matrix(y_train,y_pr_train))
```
